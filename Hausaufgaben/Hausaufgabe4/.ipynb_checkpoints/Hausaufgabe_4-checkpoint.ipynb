**b)** Wie ist der Verlauf der Geschwindigkeit des Klotzes über der Zeit? Wie der Verlauf der kinetischen und potentiellen Energie des Gesamtsystems über der Zeit? Um $v-t$ und $E_{kin}, E_{pot}-t $ Diagramme zu plotten zu können, verwende das Video, und tracke den Verlauf der Position der Feder. 

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([  ])#Werte aus dem Video : Verfolge den Mittelpunkt des Klotzes chronlogisch
t=np.array([  ])
Deltax= x_0-x

#oder vorgeben
def Deltax(t):
    return np.sin(t*np.sqrt(D/m))*x_0 

#fürs Gesamtsystem Klotz-Feder (in der Kontaktzeit)
#Ableitung
def v(t):
    return np.sqrt(D/m)*np.cos(t*np.sqrt(D/m))*x_0 

def Epot(t):
    return 0.5*D*Deltax(t)**2

def Ekin(t):
    return 0.5*m*v(t)**2

def Eges(t):
    return Epot(t)+Ekin(t)


t= np.linspace(0,np.pi*np.sqrt(m/D), 100)#s

plt.figure()
plt.title('Geschwindigkeit')
plt.plot(t, v(t))
plt.xlabel('Zeit / s')
plt.ylabel('Geschwindigkeit / m/s ')
plt.grid()

plt.figure()
plt.title('Energie')
plt.plot(t, Epot(t), 'r:' ,label='potentielle Energie')
plt.plot(t, Ekin(t),'b.' ,  label='kinetische Energie')
plt.plot(t, Eges(t),'g' , label='Gesamtenergie')
plt.xlabel('Zeit / s')
plt.ylabel('Enegie / J ')
plt.legend()
plt.grid()
```
