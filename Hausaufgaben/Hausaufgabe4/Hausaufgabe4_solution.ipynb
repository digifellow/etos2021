1.  Wie hoch ist die maximale potentielle Energie? Bestimmen sie dazu die Federkonstante und geben sie beide Werte aus.

```python
m = 2.5                     # kg
v_0 = 5                     # m/s
x_0 = 0.0                   # m
x_max = 0.25                # m
Deltax = x_max-x_0          # m
D = (m*v_0**2)/(Deltax**2)  # Nm
Epot_max = 0.5*D*Deltax**2  # J

print('Die potentielle Energie bei maximaler Stauchung ist %.2f J und die Federkonstante beträgt %.2f N/m ' %(Epot_max, D))
```

**Lösung** 

$\Delta x =x_{max}- x_0$ bezeichnet im Folgenden die maximale Ausdehnung der Feder
$$F= D \cdot \Delta x$$

$$E_{pot}= \frac{1}{2} \cdot D \cdot (\Delta x) ^2$$

$$\quad E_{kin}= \frac{1}{2} \cdot m \cdot v_0^2$$

$$\overset{\mathrm{Energieerhaltung}}{\Rightarrow}  D= \frac{m\cdot v_0^2}{(\Delta x) ^2 }$$

2. Wie der Verlauf der kinetischen, potentiellen und gesamten Energie des Gesamtsystems über der Zeit? Wie ist der Verlauf über den Ort (gemäß dem gewählten Koordinatensystem also über die Ausdehnung der Feder)? Um und $E_{kin}, E_{pot}, E_{ges}-t $ bzw. $E_{kin}, E_{pot}, E_{ges}-  x $  Diagramme plotten zu können, verwenden Sie die oberen Tabellen.         
    *Hinweis*: Bitte weisen Sie Ihre Diagramme in den angegebenen Variablen `plotn`,damit die automatische Auswertung diese finden kann.


```python
import matplotlib.pyplot as plt
import numpy as np

#Werte aus der Tabelle
t=np.array([0 , 0.0087, 0.0175, 0.0262, 0.0349, 0.0436, 0.0524, 0.0611, 0.0698, 0.0785]) 
x=np.array([0, 0.0434, 0.0855, 0.1250, 0.1607, 0.1915, 0.2165, 0.2349, 0.2462, 0.2500]) #np.sin(t*np.sqrt(D/m))*x_0 
v=np.array([5, 4.9241,  4.6985, 4.3301, 3.8302, 3.2142, 2.5002, 1.7114, 0.8685 , 0])# np.sqrt(D/m)*np.cos(t*np.sqrt(D/m))*x_0  bzw. hier über Energieerhaltung Ekinmax= Ekin+Epot für x zwischen 0 und xmax v = np.sqrt((m*v_0**2-D*x**2)/m)


# Potenzielle Energie
Epot = 0.5*D* x**2 

# Kinetische Energie
Ekin = 0.5*m* v**2

#Gesamtenergie
Eges =  Epot + Ekin


plt.figure()
plt.title('Energie über Zeit ')
plt.plot(t, Epot, 'r:' ,label='potentielle Energie')
plt.plot(t, Ekin,'b:.' ,  label='kinetische Energie')
plt.plot(t, Eges,'g' , label='Gesamtenergie')
plt.xlabel('Zeit / s')
plt.ylabel('Enegie / J ')
plt.legend()
plt.grid()

plt.figure()
plt.title('Energie über Ort')
plt.plot(x, Epot, 'r:' ,label='potentielle Energie')
plt.plot(x, Ekin,'b:.' ,  label='kinetische Energie')
plt.plot(x, Eges,'g' , label='Gesamtenergie')
plt.xlabel('Ort / m')
plt.ylabel('Enegie / J ')
plt.legend()
plt.grid()
```
